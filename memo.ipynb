```python
import pandas as pd
import urllib.request
from io import StringIO

url = "https://raw.githubusercontent.com/Yuki1005/NCUR2023_Project1_MYSHA/master/aaaaa.csv"

#csvを読み込む関数
def read_csv(url):
    print(url)
    res = urllib.request.urlopen(url)
    res = res.read().decode("shift-jis")
    df = pd.read_csv(StringIO( res) )
    return df

#実行
read_csv(url)
```
