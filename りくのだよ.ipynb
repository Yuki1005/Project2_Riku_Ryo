```python
import PySimpleGUI as sg
import pandas as pd

#https://github.com/Yuki1005/Project2_Riku_Ryo/blob/main/%E3%82%8A%E3%81%8F%E3%81%AE%E3%81%A0%E3%82%88.ipynb

sg.theme('Dark Blue 3')

radio_dic = {
    '-1-': 'A',
    '-2-': 'B',
    '-3-': 'C',
    '-4-': 'D',
    '-5-': 'E',
    '-6-': 'F',
    '-7-': 'G',
}
num = 0
bango = 0
name = list(range(1,31,1))   ##ここを項目の名前にする 


def main(radio_dic):
    global num
    global bango
    layout = []
    layout.append([sg.Text("幸せ最適化",font=("Arial",22))])
    for i in range(5):
        layout.append([sg.Text("{}".format(name[bango]))])
        layout.append([sg.Radio(item[1], key=item[0], group_id="{}".format(bango)) for item in radio_dic.items()])
        bango += 1
    if num == 5:
        layout.append([sg.Button('結果発表')])
    else:
        layout.append([sg.Button('次へ')])
    num += 1
    return sg.Window("メインウィンドウ", layout, finalize=True)

window = main(radio_dic)


kekka_list = []
while True:
    event, values = window.read()

    if event is None:
        break

    elif event == '次へ':
        if num == 6:
            for i in range(0, len(list(values.values())), 7):
                a = list(values.values())[i: i+7]
                kekka_list.append(a)
            window.close()
        else:
            for i in range(0, len(list(values.values())), 7):
                a = list(values.values())[i: i+7]
                kekka_list.append(a)
            window.close()
            window = main(radio_dic)
    elif event == "結果発表":
        for i in range(0, len(list(values.values())), 7):
                a = list(values.values())[i: i+7]
                kekka_list.append(a)
        window.close()


pd.DataFrame(kekka_list,columns=["A","B","C","D","E","F","G"]).to_csv("a.csv",index=False)
window.close()
```
